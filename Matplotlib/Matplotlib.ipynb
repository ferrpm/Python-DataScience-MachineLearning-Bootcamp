```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

### FUNCTIONAL

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'red')

plt.subplot(1,2,2)
plt.plot(y,x,'green')
```

### OBJECT ORIENTED

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) # left,bottom,width and height
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
fig,axes = plt.subplots()

axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
plt.tight_layout() # fix overlaps

# for axe in axes:
#     axe.plot(x,y)

axes[0].plot(x,y) # Create plot for the axe in position 0
axes[0].set_title('First Plot')
    
axes[1].plot(y,x) # Create plot for the axe in position 1
axes[1].set_title('Second Plot')
```

# Figure Size and DPI

```python
fig = plt.figure(figsize=(8,2))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('figure_example.png',dpi=200) # dots per pixel, more detail
```

```python
fig = plt.figure()
axe = fig.add_axes([0,0,1,1])
axe.plot(x,x**2,label='X Squared')
axe.plot(x,x**3,label='X Cubed')

# Add legends
axe.legend(loc=(0.1,0.8))
```

## Plot Appearance

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='pink', linewidth=3,alpha=1,linestyle=':',marker='o',markersize=10,  # linewidth -> lw linestyle -> ls
       markerfacecolor='red', markeredgewidth=3,markeredgecolor='blue') 
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=2,ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
